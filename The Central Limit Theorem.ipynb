# Central Limit Theorem

## Load the libraries

```python
import numpy as np
import seaborn as sns
import statistics as stat
import matplotlib.pyplot as plt
```

## Let's create a normally distributed dataset of n points

```python
norm_dist_data = np.random.normal(size=10000)
```

## And plot the data using seaborn

```python
sns.displot(norm_dist_data,kde = True)
```

## Now let's create a function to take random values of the sample and calculate the sample mean

```python
def sample_mean(dataset, sample_size, n_samples):
    sample_means = []
    for i in range(n_samples):
        sample = np.random.choice(dataset,
                                 size=sample_size,
                                 replace=False)
        sample_mean = stat.mean(sample)
        sample_means.append(sample_mean)
    return sample_means
```

## Taking 10 samples and 10 values in each

```python
norm_plot_10 = sns.displot(sample_mean(norm_dist_data,10,10),kde=True)
norm_plot_10.savefig('output.png')
```

## Let's take more values in the same number of samples

```python
norm_plot_1000 = sns.displot(sample_mean(norm_dist_data,10,1000),kde=True)
norm_plot_1000.savefig('norm_1000.png')
```

As we can see the distribution is now looking normally-distributed

```python
norm_plot_1000 = sns.displot(sample_mean(norm_dist_data,1000,10),kde=True)
norm_plot_1000.savefig('norm_10_1000.png')
```

# Let's also increase the number of samples

```python
norm_dist_1000_1000 = sns.displot(sample_mean(norm_dist_data,1000,1000),kde=True)
norm_dist_1000_1000.savefig('norm_1000_1000.png')
```

## The samples are coming from a normally distributed dataset, what would happen if it wasn't normally distributed?<br>
## Let's use a skewed dataset

```python
from scipy.stats import skewnorm
```

```python
sk_dist_data = skewnorm.rvs(12,size=10000)
```

## Explore the data

```python
skew = sns.displot(sk_dist_data,kde=True)
skew.savefig('skew.png')
```

## The distribution is not normally-distributed looking.<br>
## Let's take more values from the dataset

## It doesn't matter where does the data come from, the median sample is going to have a normal distribution as the number of sample increases

```python
mul_dist_data = np.concatenate((np.random.normal(size=10000), np.random.normal(loc = 4.0, size=10000)))
```

```python
sns.displot(mul_dist_data,kde=True)
```

## But what is the correct number of samples to obtain a normal distribution?

```python
_ = sns.displot(sample_mean(mul_dist_data,500,10),kde=True)
```

```python
_ = sns.displot(sample_mean(mul_dist_data,500,20),kde=True)
```

```python
_ = sns.displot(sample_mean(mul_dist_data,500,30),kde=True)
```

```python
_ = sns.displot(sample_mean(mul_dist_data,500,100),kde=True)
```

## Now let's use a Pareto Distribution (which variance is infinite) and test it's behaviour

```python
par_dist = np.random.pareto(10000,size=10000)
```

```python
sns.displot(par_dist,kde = True)
```

## Not even with 1000 samples/values we are getting a normal distribution, remember one of the requirementes for the C.L.T. is being able to get the mean and variance

```python
par_dist_img = sns.displot(sample_mean(par_dist,1000,1000),kde=True)
par_dist_img.savefig('par.png')
```
